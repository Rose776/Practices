```python

#First, let's import the necessary libraries for generating random numbers and plotting the results.
import numpy as np
import matplotlib.pyplot as plt

#Next, we will define a function to generate a random sample from a uniform distribution. 
#This sample will be used to calculate the sample mean and standard deviation.

def random_sample(n):
    return np.random.uniform(0,1,n)
#Now, we will take random samples from the population and calculate their sample mean and standard deviation. 
#We will also store the sample means and standard deviations in two separate lists.

sample_means = []
sample_stds = []
for i in range(1,1000):
    
    sample = random_sample(i)
    sample_means.append(np.mean(sample))
    sample_stds.append(np.std(sample))
    
## Plot the results

plt.plot(sample_means, label='Sample Mean')
plt.plot(sample_stds, label='Sample Standard Deviation')
plt.axhline(0.5, color='r', label='Population Mean')
plt.axhline(0.2887, color='g', label='Population Standard Deviation')
plt.title('Central Limit Theorem')
plt.xlabel('Sample Size')
plt.ylabel('Mean & Standard Deviation')
plt.legend()
plt.show()
```
